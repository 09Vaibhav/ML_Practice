```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/practice/haberman.csv")
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df.columns = ["age","year","l_node","survival_status"]
```

```python
print(df)
```


*   1 - PATIENT SURVIVED MORE THAN 5 YEARS AFTER OPERATION
*   2 - PATIENT SURVIVED LESS THAN 5 YEARS AFTER OPERATION






```python
print(df.info())   #GIVE INFO ABOUT NULL OR MISSING DATA AND DATA TYPE
```

```python
df["survival_status"] = df["survival_status"].map({1: 'yes',2:'no'})   #FOR UNDERSTANDING REPLACING 1,2 BY YES/NO
```

```python
print(df)
```

```python
df.describe()
```

```python
df["survival_status"].value_counts()    #COUNTS THE TOTAL NUMBER OF EACH CLASS(YES/NO)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
status_yes = df.loc[df["survival_status"]== 'yes']    
status_no = df.loc[df["survival_status"]== 'no']
```

```python
status_yes.describe()
```

**UNIVARIATE DATA ANALYSIS(HISTOGRAM ,PDF,CDF)**

```python
sns.set_style('whitegrid')
m=sns.FacetGrid(df,hue="survival_status",height= 4)
m.map(sns.distplot,'age')
m.add_legend()
plt.show()

#those who are in age between 30-40 have high chances of survival
```

```python
g = sns.FacetGrid(df,hue='survival_status',size=5)
g.map(sns.distplot,'year')
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(df,hue='survival_status',size=5)
g.map(sns.distplot,'l_node')
g.add_legend()
plt.show()

#here patient who have lymph_node of 1 have high chances of survival (more than 5 years after operation)
```

```python
counts1,bin_edges1 = np.histogram(status_yes['l_node'],bins=10,density=True)
pdf1 = counts1/(sum(counts1))
```

```python
#cdf
cdf1= np.cumsum(pdf1)
plt.plot(bin_edges1[1:],pdf1)
plt.plot(bin_edges1[1:],cdf1,label = 'yes')
plt.xlabel("l_node")
plt.ylabel("probability")
counts2,bin_edges2 = np.histogram(status_no['l_node'],bins=10,density=True)
pdf2 = counts2/(sum(counts2))
cdf2= np.cumsum(pdf2)
plt.plot(bin_edges2[1:],pdf2)
plt.plot(bin_edges2[1:],cdf2,label = 'no')
plt.xlabel("l_node")
plt.ylabel("probability")
plt.legend()
plt.show()
#less node, more chances of survival
```

```python
sns.boxplot(x='survival_status',y='age',data =  df)
plt.show()
sns.boxplot(x='survival_status',y='year',data =  df)
plt.show()
sns.boxplot(x='survival_status',y='l_node',data =  df)
plt.show()
```

```python
sns.violinplot(x='survival_status',y='age',data =  df)
plt.show()
sns.violinplot(x='survival_status',y='year',data =  df)
plt.show()
sns.violinplot(x='survival_status',y='l_node',data =  df)
plt.show()
```

from 3rd plot, we see that for l_node in between 0-2 , the chances of survival status is high
